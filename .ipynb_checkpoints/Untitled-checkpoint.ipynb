```python
import json
```

```python
with open('qa.json') as f:
    qa = json.load(f)
```

```python
x = qa[0].get('QuestionAnswer')
```

```python
def divide_chunks(l, n): 
      
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 
```

```python
out = ['<!DOCTYPE html>']
for k, e in enumerate(qa):
    x = e.get('QuestionAnswer')
    
    out.append('<h1>'+x[0]+'</h1>')
    y = divide_chunks(x[1:], 9)
    for t in y:
        out.append(''.join('<p>'+'<br>'.join(t)+'</p>'))

with open('index2.html', 'w') as f:
    for i in out:
        f.write(i)
```
